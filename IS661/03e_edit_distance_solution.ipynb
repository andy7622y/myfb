````python
from nltk.corpus.reader.util import StreamBackedCorpusView

# solution
def clean(tokens: Union[List[str], StreamBackedCorpusView]) -> List[str]:
    """
    :param tokens: list of tokens that should be cleaned
    :return: list of cleaned tokens
    """
    stop_words = stopwords.words('english')
    punct = string.punctuation

    clean_tokens = [t for t in tokens if not (t.lower() in stop_words or t.lower() in punct)]

    return clean_tokens
````

````python
def clean1(tokens: Union[List[str], StreamBackedCorpusView]) -> List[str]:
    """
    :param tokens: list of tokens that should be cleaned
    :return: list of cleaned tokens
    """

    removals = stopwords.words('english')
    removals.extend(string.punctuation)
    return [i for i in word_tokenize(tokens) if i.lower() not in removals]
````

### Task 2: Levenshtein distance algorithm

__a)__ In this task we want to calculate the minimum edit distance between two input strings.

Write a function that implements the minimum edit distance algorithm (dynamic programming) from the lecture. Your function should get two strings as input and should return the minimum edit distance and a 2-dimensional numpy array that represents the corresponding edit distance table ``D``. Assume that insertion and deletion have a cost of 1. The cost of substitution should be specified by the user (default: 2).

Note, that the entry ``D[i][j]`` in the returned numpy array should correspond to the cell D(i,j) of the edit distance table from the lecture (i.e. the numpy array is a vertically flipped version of the table from the lecture).

__Example:__ _The edit distance table_ ``D`` _for a source string of length 3 and a target string of length 5 should be initialized as follows:_

```
[[0  1  2  3  4  5]
 [1  0  0  0  0  0]
 [2  0  0  0  0  0]
 [3  0  0  0  0  0]]
```

````python
def edit_distance(str1: str, str2: str, cost_of_substitute: int=2) -> (int, NDArray[NDArray[int]]):
    '''
    :param str1: source string
    :param str2: target string
    :cost_of_substitute: cost for substituting mismatching letters
    :returns:
        - minimum edit distance of str1 and str2
        - minimum edit distance table
    '''
    I = len(str1) + 1  # +1 for hashtag symbol (empty string)
    J = len(str2) + 1
    D = np.zeros((I,J), dtype=int)

    # initialize first row and column
    D[:,0] = np.arange(I)  # index 0 of all elements (first column)
    D[0,:] = np.arange(J)  # all entries of element at index 0 (first row)

    for i in range(1,I):  # loop through rows
        for j in range(1,J):  # loop through columns
            sub = D[i-1, j-1] # init cost same letters
            if str1[i-1] != str2[j-1]: # input strings do not include hashtag symbol (empty string) => index-1 otherwise will exceed the length
                  sub += cost_of_substitute # not same letters

            D[i,j] = min(D[i-1,j] + 1, D[i,j-1] + 1, sub)  # new value

    return D[-1][-1], D
    # return D[I-1][J-1], D
````

````python
dist, D = edit_distance("sitting", "kitten")
print(dist)
print(D)
````
