Today we will be exploring **The Titanic dataset** which is commonly used in introductory Machine Learning classes.
It contains data for 887 passengers on the Titanic. Each row represents one person. The columns describe different attributes about the person including whether they survived, their age, their on-board class, their sex, and the fare they paid.
This dataset is already included to seaborn package

```python
import seaborn as sns
sns.set_theme()
titanic = sns.load_dataset('titanic');
titanic.head()
```

# For the three tasks below try to use as many vizualization as possible

## Plot the distribution of passengers age and fare

```python
sns.displot(data = titanic, x = 'fare')
```

```python
sns.histplot(data = titanic, x = 'fare', bins=20)  # sns.histplot has an optional parameter 'bins'
```

```python
sns.histplot(data = titanic, x = 'age', bins=20)
```

```python
sns.histplot(data = titanic, x = 'age', y = 'fare', bins=20)  # make a 2D histplot
```

```python
sns.jointplot(data = titanic, x = 'age', y = 'fare', s=10)  # sns.jointplot is an alternative that can combine e.g. scatter- & histplot
```

## Check how these distributions vary depending on gender of passengers

```python
sns.jointplot(data = titanic, x = 'age', y = 'fare', hue = 'sex')  # add attribute to previous plot
```

```python
sns.boxplot(data = titanic, x = 'sex', y = 'age', hue='sex')  # we have a categorical variable, so we can use e.g. sns.boxplot
```

```python
sns.stripplot(data = titanic, x = 'sex', y = 'fare', hue='sex', size=3, jitter=0.3)  # another option to illustrate 'fare'
```

## Check how these distributions vary depending on the class variable

```python
sns.boxplot(data = titanic, x = 'class', y = 'fare', hue='class')
```

```python
sns.histplot(data = titanic, x = 'fare', hue = 'class', bins=20, multiple='stack')
```

# Visually explore the dataset and try to find what variables are closely related to survival chance. What criteria might be used to classify passengers as alive or not?

```python
sns.barplot(data = titanic, y = 'age', x = 'sex', hue = 'alive')  # age and sex might be indicators
```

```python
# Plot the distributions separately for male and female
sns.displot(data = titanic[titanic.sex == 'male'], x = 'age', hue = 'alive', kind="hist")
```

```python
sns.displot(data = titanic[titanic.sex == 'female'], x = 'age', hue = 'alive', kind="hist")
```

```python
sns.displot(data = titanic, x = 'fare', hue = 'alive', kind="kde",  # KDE can help visualizing multiple overlapping histograms but is less precise
            bw_adjust=.5)  # adjust the bandwidth (how much smoothing is applied)
```

```python
# This a (helpful) brute-force approach to finding patterns
# Note that some categorical variables such as 'sex' are not included!
sns.pairplot(titanic, hue = 'alive', corner=True)
```
