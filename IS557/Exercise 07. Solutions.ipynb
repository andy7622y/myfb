## Swap the first and the third rows of this array

```python
# store the third row into a temporary buffer
# make sure to use copy(), otherwise the buffer will just reference x
buffer = x[2,:].copy()  # x[2] is also possible
buffer
```

```python
# replace the third row with the first row
x[2,:] = x[0,:]
# replace the first row with the buffer
x[0] = buffer
x
```

```python
# we can also do this in one step by selecting both rows at once
x[[0,2],:] = x[[2,0],:]
x  # first and third row are now switched back
```

## Create a 8x8 array to represent a chess board, such as that 0s correspond to white squares and 1s to black squares

```python
# create a board of zeros first
x = np.zeros((8,8))

# add ones to all odd entries in every even row
x[1::2, ::2] = 1
x
```

```python
# finally, add ones to all even entries in every odd row
x[::2, 1::2] = 1
x
```

## Ploting trajectory
If an object is thrown into the air with a velocity $v$ and angle $\alpha$ from horizontal then its trajectory can be computed using the following formula:

$y = x\tan\alpha - \frac{gx^2}{2v^2\cos^2\alpha}$.

Plot the trajectory of an object for a specified $v$ and $\alpha$, experiment with different values of $\alpha$ by ploting multiple trajectories on the same plot.

Hint: np.linspace(start, end, N) function creates a numpy array with N numbers equally spaced between start and end

```python
# global constant
g = 9.81

# calculate the trajectory given v and alpha
def get_trajectory(v, alpha):
    
    # how far does it fly?
    max_x = np.tan(alpha) / g * (2 * v * v * np.cos(alpha) ** 2)
    
    # create 50 waypoints
    x = np.linspace(0, max_x)
    
    # calculate the height at each waypoint using the given formula
    y = np.tan(alpha) * x - g * x ** 2 / (2 * v * v * np.cos(alpha) ** 2)
    
    return x, y
```

```python
import matplotlib.pyplot as plt
```

```python
# fixed velocity
v = 10

# plot some trajectories
for alpha in range(0, 90):
    x, y = get_trajectory(v, alpha / 180 * np.pi) # np.tan uses radian angles
    plt.plot(x, y, color='blue', alpha=0.2)

# highlight some specific angles
alpha = 45
x, y = get_trajectory(v, alpha / 180 * np.pi)
plt.plot(x, y, color='red')

alpha = 30
x, y = get_trajectory(v, alpha / 180 * np.pi)
plt.plot(x, y, color='orange')

alpha = 60
x, y = get_trajectory(v, alpha / 180 * np.pi)
plt.plot(x, y, color='purple')

plt.show()
```
