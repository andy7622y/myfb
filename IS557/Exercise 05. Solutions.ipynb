## Working with files

Create a function called copy_file that takes two arguments: the source file path and the destination file path. The function should read the content of the source file line by line and write it to the destination file.

```python
def copy_file(in_name, out_name):
    with open(in_name, 'r') as f_in:
        with open(out_name, 'w') as f_out:
            for line in f_in:
                f_out.write(line)

copy_file('textfile.txt', 'textfile2.txt')
```

Write a function that reads the content of a text file, encrypts it using a simple Caesar cipher with a user-specified shift value, and saves the encrypted content to a new file. Write another function that reads the encrypted file, decrypts the content using the shift value, and prints the decrypted content.

*Caesar cipher â€” is a type of substitution cipher in which each letter in the plaintext is replaced by a letter some fixed number of positions down the alphabet. For example, with a shift value of 3, A would be replaced by D, B would become E, and so on.*

```python
latin = 'abcdefghijklmnopqrstuvwxyz'
LATIN = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

def do_shift(string, shift):
    result = ''
    for c in string:
        if c in latin:
            ind = latin.index(c)
            ind = (ind + shift) % 26
            result += latin[ind]
            
        elif c in LATIN:
            ind = LATIN.index(c)
            ind = (ind + shift) % 26
            result += LATIN[ind]
            
        else:
            result += c
            
    return result
```

```python
do_shift('A, b, c, z', 1)
```

```python
def encrypt(file_name, shift):
    with open(file_name + '.txt', 'r') as f:
        text = f.read()

    new_text = do_shift(text, shift)

    with open(file_name + '_secret.txt', 'w') as f:
        f.write(new_text)


def decrypt(file_name, shift):
    with open(file_name + '.txt', 'r') as f:
        text = f.read()

    decr = do_shift(text, -shift)
        
    print(decr)
```

```python
encrypt('textfile', 11)
```

```python
with open('textfile_secret.txt') as f:
    print(f.read()[:990])
```

```python
decrypt('textfile_secret', 11)
```

# Datetime
Write a function to determine the time difference between two timezones.

```python
import datetime
import pytz

def timediff(time_zone_1, time_zone_2):
    time_now = datetime.datetime.now()
    
    t_1 = time_now.astimezone(pytz.timezone(time_zone_1))
    t_2 = time_now.astimezone(pytz.timezone(time_zone_2)) 
    
    time_1 = t_1.hour + t_1.minute / 60
    time_2 = t_2.hour + t_2.minute / 60
    
    result = time_1 - time_2
    
    if result < -12:
        result += 24.0
    
    if result > 12:
        result -= 24.0
    
    return result

print(timediff('Pacific/Chatham','Europe/Berlin'))
print(timediff('Asia/Katmandu','Europe/Berlin'))
```

Write a function to determine the number of days between today and your next birthday

```python
def timetobirthday(month, day):
    t_d = datetime.date.today()
    b_d = datetime.date(t_d.year, month, day)
    
    if b_d < t_d:
        b_d = datetime.date(t_d.year + 1, month, day)
        
    return (b_d - t_d).days
```

```python
timetobirthday(1, 26)
```

Write a function that lists the dates of all Sundays of a given year

```python
def allSundays(year):
    start_date = datetime.date(year, 1, 1)
    
    first_sunday = start_date + datetime.timedelta(6 - start_date.weekday())
    
    print(first_sunday)
    
    next_sunday = first_sunday

    while True:
        next_sunday += datetime.timedelta(7)
        if next_sunday.year != year:
            break
        print(next_sunday)
```

```python
allSundays(2023)
```

Write a Python function to count the number of times Friday was the 13st day of the month in a given year

```python
def friday13(year):
    for i in range(1, 13):
        dt = datetime.date(year, i, 13)
        if dt.weekday() == 4:
            print(dt, end = ' ')
```

```python
for year in range(2000, 2031):
    friday13(year)
    print()
```

## Random numbers

Write a function that generates a random password with a user-specified length. The password should contain a mix of uppercase letters, lowercase letters, digits, and special characters. Ensure that the generated password contains at least one character from each category.

*Hint: you might want to use [string](https://docs.python.org/3/library/string.html) module*

```python
import random
import string

def random_char(string):
    return string[random.randint(0, len(string) - 1)]
```

```python
upper = string.ascii_uppercase
lower = string.ascii_lowercase
digits = string.digits
special = string.punctuation
```

```python
def get_password(length):
    if length < 4:
        print('Length should be at least 4 characters')
        return
    
    password = random_char(upper) + random_char(lower) + random_char(digits) + random_char(special)  # add one from each category
    
    all_ = upper + lower + digits + special
    
    for i in range(length - 4):
        password += random_char(all_)  # add more random character from any category
        
    list_ = list(password)
    random.shuffle(list_)  # shuffle to make sure that the categories of the first 4 characters cannot be guessed
    password = ''.join(list_)
    
    return password
```

```python
get_password(16)
```
