Then this could be complicated by introducing uncertainly. For example, instead of 7% select a random (but realistic) return for each year, do a lot of simulations (e.g. 10,000) and plot the results for the worst and the best outcome.

```python
def money_now(m):
  annual_return = 0.07
  money = 100
  monthly_return = (1 + annual_return) ** (1/12) - 1
  for i in range(1,m+1):
    money = money * (1 + monthly_return)

  return money
```
