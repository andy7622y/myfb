## Selecting Parts of the DataFrame

```python
df2 = df[['gender', 'language', 'has_mother']]  # get the specified columns – returns a DataFrame
df2
```

We can select rows with `iloc` and `loc`. `iloc` is integer position-based and `loc` is label-based.

```python
df.iloc[5]  # select the 6th row – returns a Series
```

```python
df2 = df.iloc[5:15]  # copy the 6th to 15th row into a new DataFrame
df2  # notice how the index stays the same
```

Why do `loc` and `iloc` behave differently in the following example?

```python
print(df2.iloc[5].depression)
print(df2.loc[5].depression)
```

```python
df2.reset_index(inplace = True)  # this helps, but it creates a new column 'index'
df2
```

```python
df3 = df.iloc[5:15]
df3.reset_index(inplace = True, drop = True)  # fully reset the index
df3
```

### Applying a function
Let's define a 'disadvantage' metric that is slightly more complicated:

```python
def disadvantaged(row):
    ind = 0
    if not row['has_grandmother']:
        ind += 0.5
    if not row['has_father']:
        ind += 1
    if not row['has_mother']:
        ind += 2
    if not row['language']:
        ind += 1
    if row['nationality'] != 25:
        ind += 0.5
        
    if row['city'] <= 3:
        ind +=1
        
    if row['city'] <= 6:
        ind += 0.5
        
    return ind
```

```python
df["disadvantage"] = df.apply(disadvantaged, axis = 1)  # apply the above defined function to create a new column
df.head()
```

```python
import seaborn as sns  # plot the results with Seaborn – see next lecture

sns.boxplot(x='disadvantage', y='math_performance', data = df)
sns.despine()
```

```python
df['is_depressed_2'] = df.apply(lambda x: 1 if x['depression'] >= 10 else 0, axis=1)  # we can also use Lambda functions
df.sample(10)
```

## Concatenate and Join DataFrames

There several methods to bring data from different dataset together, most importantly `concat` and `merge`.

```python
df1 = pd.DataFrame ([["Student 1", 2.0, 1.5],
                     ["Student 2", 3.0, 2.5],
                     ["Student 3", 1.5, 2.0],
                     ["Student 4", 1.0, 4.0]], columns = ["Student", "course_A", "course_B"])


df2 = pd.DataFrame ([["Student 5", 2.3, 2.5],
                     ["Student 6", 3.0, 1.0],
                     ["Student 7", 2.5, 2.0],
                     ["Student 8", 1.7, 3.5]], columns = ["Student", "course_A", "course_B"])
df1.head(), df2.head()
```

```python
df_all = pd.concat([df1,df2])  # concatenate the DataFrames on axis 0 (index) – merges entries according to column names
df_all
```

```python
df_all.reset_index(inplace=True, drop=True)  # after concatenating, we need to adjust the index
df_all
```

```python
df1 = pd.DataFrame ([["Student 1", 2.0, 1.5],
                     ["Student 2", 3.0, 2.5],
                     ["Student 3", 1.5, 2.0],
                     ["Student 4", 1.0, 4.0]], columns = ["Student", "course_A", "course_B"])


df2 = pd.DataFrame ([["Student 2", 2.3, 2.5],
                     ["Student 6", 3.0, 1.0],
                     ["Student 7", 2.5, 2.0],
                     ["Student 8", 1.7, 3.5]], columns = ["Student", "course_C", "course_D"])
```

```python
df_all = pd.concat([df1,df2], ignore_index = True)  # notice how 'NaN' is added wherever no value is available
df_all
```

```python
df1 = pd.DataFrame ([["Student 1", 2.0, 1.5],
                     ["Student 2", 3.0, 2.5],
                     ["Student 3", 1.5, 2.0],
                     ["Student 4", 1.0, 4.0]], columns = ["Student", "course_A", "course_B"])


df2 = pd.DataFrame ([["Student 2", 2.3, 2.5],
                     ["Student 3", 3.0, 1.0],
                     ["Student 7", 2.5, 2.0],
                     ["Student 8", 1.7, 3.5]], columns = ["Student2", "course_C", "course_D"])
df1, df2
```

```python
df1.merge(df2, left_on = 'Student', right_on = 'Student2', how = 'inner')  # merge and only keep students that are in both DataFrames
```

```python
df1.merge(df2, left_on = 'Student', right_on = 'Student2', how = 'outer')  # merge and keep all students (fill with 'NaN' if necessary)
```
