# Matplotlib
Matplotlib is the most widely used plotting library in Python. It is relatively straightforward to begin with, but creating high-quality, publication-ready plots can be more challenging. Although there are other libraries that can produce higher quality plots straight away, many are built upon matplotlib, making it essential to learn.

```python
import matplotlib.pyplot as plt
```

To start plotting, you simply need to pass a list (or lists) containing the data to the plot function.

```python
x = []
y = []
with open('Lecture 06.data.csv', 'r') as f:  # load some data from a file
    for line in f:
        data = line.strip()
        x.append(float(data.split(',')[0]))
        y.append(float(data.split(',')[1]))
```

```python
x[:10]
```

The command below will plot the values contained in the list `x`. If only one list is passed, it is assumed that these are the values for the y-axis, and the x-axis values will range from 0 to the length of the list minus one.

```python
plt.plot(x)
```

You can also plot multiple graphs on the same figure.

```python
plt.plot(x)
plt.plot(y)
```

Apart from line plots, matplotlib also supports other types of plots, such as scatter plots, histograms, bar charts, and pie charts.

```python
plt.scatter(x, y)
```

```python
plt.hist(x)
```

```python
help(plt.hist)
```

```python
plt.hist(x, 100)
```

```python
plt.bar(['first element', 'last element'], [x[0], x[-1]])
```

```python
plt.pie([x[0], x[-1]], labels = ["first element", "last element"])
```

Simple plots created with plt.plot() are already useful for gaining a better understanding of your data. However, such plots are not suitable for publication. There are numerous parameters and additional functions that allow you to customize your plot's appearance. Preparing publication-ready plots is usually a tedious process, often involving trial and error and referring to the documentation or online resources.

```python
# Let's specify figure size, squares are usually better
plt.figure(figsize = (4, 4))
# Let's increase the width of our lines (almost always better), and add labels
plt.plot(x, linewidth = 3, label = 'NVIDIA')
plt.plot(y, lw = 3, label = 'Starbucks')
#Now let's label axes, and set the font size size
fs = 16
plt.xlabel('days', size = fs)
plt.ylabel('price, $', size = fs)
plt.title('Stock price', size = fs)
# Now let's specify ticks
plt.xticks([0, 500, 1000, 1500], size = fs)
plt.yticks([0,100, 200, 300, 400], size = fs)
# Y axis should start from zero
plt.ylim(0, 500)
# Let's add legend and remove border
plt.legend(frameon = False, prop = {'size': 16})
```

Now let's make further improvments

```python
plt.figure(figsize = (4,4))

plt.plot(x, label = 'NVIDIA', lw = 3, color = '#76B900')
plt.plot(y, label = 'STARBUCKS', lw = 3, color = '#00704A')
plt.ylabel('stock price, $', size = 18)
plt.legend(prop = {'size': 16}, frameon = False)
plt.ylim(0, 450)
plt.xticks([0, 365, 730, 1095, 1460],
           ['Apr 2019', 'Apr 2020', 'Apr', 'Apr 2022', 'Apr 2023'],
           size = 12, rotation = 45)

plt.yticks([0, 100, 200, 300, 400], size = 14)

ax = plt.gca()
ax.spines['top'].set_color('None')
ax.spines['right'].set_color('None')
```
